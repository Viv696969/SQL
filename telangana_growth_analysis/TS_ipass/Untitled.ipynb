```python
from pandas import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=read_csv("top_sectors_by_invt_done_in_2022.csv")
```

```python
fig=df.set_index("sector").plot.bar(color='red')
plt.ylabel("total investment in crores")
plt.title("top sectors with highest investment in 2022")
plt.xticks(rotation=45, ha='right', fontsize=8)
# plt.legend(loc=4)
plt.tight_layout()
fig.bar_label(fig.containers[0])
plt.savefig("top_sectors_by_invt_done_in_2022.png",dpi=300)
```
