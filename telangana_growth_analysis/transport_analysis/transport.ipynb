```python
from pandas import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=read_csv("total_veh_sales_by_month.csv")
```

```python
df.set_index("Mmm").plot.bar(color='red')
plt.xlabel("Months")
plt.ylabel("total sales in Million")
plt.title("Total Vehicle sales by month")
plt.tight_layout()
plt.savefig("total_veh_sales_by_month.png",dpi=300)
```

```python
df=read_csv("total_veh_sales_by_quarter.csv")
```

```python
df.set_index("quarter").plot.bar(color='red')
# plt.xlabel("Q")
plt.ylabel("total sales in Million")
plt.title("Total Vehicle sales by quarter")
plt.tight_layout()
plt.savefig("total_veh_sales_by_quarter.png",dpi=300)
```

```python
df=read_csv("top_districts_for_motor_cycle.csv")
```

```python
df.set_index("district").plot.bar()
plt.ylabel("total sales in million")
plt.title("top districts with motor cycle sold")
plt.tight_layout()
plt.savefig("top_districts_for_motor_cycle.png",dpi=300)
```

```python
read_csv("top_districts_for_motor_car.csv").set_index("district").plot.bar()
plt.ylabel("total sales in million")
plt.title("top districts with motor cars sold")
plt.tight_layout()
plt.savefig("top_districts_for_motor_car.png",dpi=300)
```

```python
read_csv("top_districts_for_motor_rikshaw.csv").set_index("district").plot.bar(color='yellow')
plt.ylabel("total sales in million")
plt.title("top districts with rikshaws sold")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.savefig("top_districts_for_motor_rikshaw.png",dpi=300)
```

```python
read_csv("top_districts_for_motor_agri.csv").set_index("district").plot.bar(color='yellow')
plt.ylabel("total sales in million")
plt.title("top districts with agriculture motors sold")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.savefig("top_districts_for_motor_agri.png",dpi=300)
```

```python
read_csv("top_districts_2022_gt_2021.csv").set_index("district").plot.bar()
plt.ylabel("total sales in million")
plt.title("top districts with more sales in 2022 than 2021")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.legend(loc=4)
plt.tight_layout()
plt.savefig("top_districts_2022_gt_2021.png",dpi=300)
```

```python
read_csv("top_districts_2021_gt_2022.csv").set_index("district").sort_values("total_2021").plot.bar()
plt.ylabel("total sales in million")
plt.title("top districts with more sales in 2021 than 2022")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.legend(loc=4)
plt.tight_layout()
plt.savefig("top_districts_2021_gt_2022.png",dpi=300)
```

```python
read_csv("bottom_districts_2022_gt_2021.csv").set_index("district").sort_values("total_2022",).plot.bar()
plt.ylabel("total sales in million")
plt.title("bottom 3 districts with more sales in 2022 than 2021")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.legend(loc=4)
plt.tight_layout()
plt.savefig("bottom_districts_2022_gt_2021.png",dpi=300)
```
