```python
telanaga_cities=[
    'Adilabad',
    'Peddapalle',
    'Karimnagar',
    'Nizamabad',
    'Zahirabad',
    'Medak',
    'Malkajgiri',
    'Secundrabad',
    'Hyderabad',
    'CHEVELLA',
    'Mahbubnagar',
    'Nagarkurnool',
    'Nalgonda',
    'Bhongir',
    'Warangal',
    'Mahabubabad',
    'Khammam',
]
```

```python
df['dummy']=df.apply(lambda row:row['state']+'a',axis=1)
```

```python
df
```

```python
df.drop('dummy',axis=1,inplace=True)
```

```python
df
```

```python
def change_state(row):
    if row['pc_name'] in telanaga_cities:
        return "Telangana"
    return row['state']
```

```python
df['state_transformed']=df.apply(lambda row:change_state(row),axis=1)
```

```python
df
```

```python
df.state_transformed.unique()
```

```python
df['state']=df.state_transformed
```

```python
df.drop('state_transformed',axis=1,inplace=True)
```
