```python
prev_states=list(df_pc.state_id.unique())
```

```python
curr_states=list(df_1.state_id.unique())
```

```python
data={'state_id':[],'pc_name':[],"pc_id":[]}
```

```python
df_1['pc_name']=df_1.pc_name.str.title()
```

```python
for i in curr_states:
    if i not in prev_states:
        print(i)
        for j in df_1[df_1.state_id==i]['pc_name'].unique():
            data['state_id'].append(i)
            data['pc_name'].append(j)
            pc_id+=1
            data['pc_id'].append(pc_id)
    else:
        temp=list(df_1[df_1.state_id==i].pc_name.unique())
        for j in temp:
            if j not in list(df_pc[df_pc.state_id==i]['pc_name'].unique()):
                data['state_id'].append(i)
                data['pc_name'].append(j)
                pc_id+=1
                data['pc_id'].append(pc_id)
```
