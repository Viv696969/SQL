```python
def change_state(row):
    if row['pc_name'] in telanaga_cities:
        return "Telangana"
    return row['state']
```

```python
df['state_transformed']=df.apply(lambda row:change_state(row),axis=1)
```

```python
states=[]
pc_names=[]
for state,grouped_state_df in df.groupby('state'):
    pc_names_unique_temp=list(grouped_state_df['pc_name'].unique())
    state_temp=[state]*len(pc_names_unique_temp)
    states+=state_temp
    pc_names+=pc_names_unique_temp
```
